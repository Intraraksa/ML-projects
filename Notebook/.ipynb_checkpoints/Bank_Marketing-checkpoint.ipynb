```python
def age_seg(age):
    if 0<age<=30:
        gen = 1
    elif 30<age<=50:
        gen = 2
    elif 50<age<=70:
        gen = 3
    else:
        gen = 4
    return gen
```

```python
data['Gen'] = data['Age'].apply(age_seg)
```

```python
sns.violinplot(x = 'Gen',y = 'Spending Score (1-100)',data = data,hue = 'Gender')
```

```python
sns.violinplot(x = 'Gen',y = 'Annual Income (k$)',data = data,hue = 'Gender')
```

```python
sns.swarmplot(x = 'Gen',y = 'Spending Score (1-100)',data = data,hue='Gender')
```
