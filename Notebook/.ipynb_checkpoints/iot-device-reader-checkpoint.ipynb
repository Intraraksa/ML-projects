```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import datetime as dt
%matplotlib inline
```

```python
data = pd.read_csv('../temperature-readings-iot-devices/IOT-temp.csv',date_parser = ['noted_date'])
```

```python
data['noted_date'] = pd.to_datetime(data['noted_date'])
```

```python
data['Date'] = data['noted_date'].dt.day
data['year'] = data['noted_date'].dt.month
data['time'] = data['noted_date'].dt.time
```

```python
df = data[data['Date']== 12]
```
