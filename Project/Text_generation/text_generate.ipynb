```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
nltk.download('punkt')
```

```python
with open('moby_dick_four_chapters.txt') as file:
    texts = file.read()
```

```python
texts
```

```python
def text_process(text):
#   word = text.lower()
#   word = word.split()
    word = re.sub('[^A-za-z1-9]',' ',text)
    word = word.lower()
    word = word.split()
#     word = [words for words in word if words not in stopwords.words('english')]
#     word = [SnowballStemmer('english').stem(words) for words in word]
    word = ' '.join(word)
    return word
```

```python
text_cleaned = text_process(texts)
```

```python
token = word_tokenize(text_cleaned)
```

```python
length = len(token)
```

```python
words = list(set(token))
```

```python
##create Word to index
word_idx = {}
for i,j in enumerate(words):
    word_idx[j]=i
```

```python
## Create Index to word
idx_word = {}
num = 0
for i in word_idx:
    idx_word[num] = i
    num = num+1
```

```python
word_to_generated = []
word_inputs = []
word_thredhold = 10
start_idx = 0
for i in range(word_thredhold,length):
    word_inputs.append(token[start_idx:i+1])
#     word_to_generated.append(token[i+1-word_thredhold:])
    start_idx = start_idx+1
```

## Create X , y

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(text_cleaned.split())
num_word = len(tokenizer.word_index)
seq = tokenizer.texts_to_sequences(word_inputs)
```

```python
y = np.array(seq)[:,-1]
y = tf.keras.utils.to_categorical(y,num_classes=num_word+1)
y
```

## Create Model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Embedding(num_word,256,input_length=11),
                                   tf.keras.layers.LSTM(150,return_sequences=True),
                                   tf.keras.layers.LSTM(150),
                                   tf.keras.layers.Dense(150),
                                   tf.keras.layers.Dense(num_word+1,activation='softmax')])
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
             loss=tf.keras.losses.categorical_crossentropy,
             metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, y, batch_size=128, epochs=300,verbose=1)
```

```python
def text_process2(text):
#   word = text.lower()
#   word = word.split()
    word = re.sub('[^A-za-z1-9]',' ',text)
    word = word.lower()
    word = word.split()
#     word = [words for words in word if words not in stopwords.words('english')]
#     word = [SnowballStemmer('english').stem(words) for words in word]
    # word = ' '.join(word)
    return word
```

```python
def generate_text(model,tokenizer,seq_length,seed_text,num_of_gen):
  text_generated = text_process2(seed_text)

  for i in range(num_of_gen):
    encoding_text = tokenizer.texts_to_sequences(text_generated)[-1]

    pad_encode = pad_sequences([encoding_text],seq_length)

    predict_seq = model.predict(pad_encode)
    predict_text = np.argmax(predict_seq,axis=1)
    text_generated.append(tokenizer.index_word[predict_text[0]])
  return text_generated
```
