### Context

This classic dataset contains the prices and other attributes of almost 54,000 diamonds. It's a great dataset for beginners learning to work with data analysis and visualization.

Content
price price in US dollars (\$326--\$18,823)

carat weight of the diamond (0.2--5.01)

cut quality of the cut (Fair, Good, Very Good, Premium, Ideal)

color diamond colour, from J (worst) to D (best)

clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x length in mm (0--10.74)

y width in mm (0--58.9)

z depth in mm (0--31.8)

depth total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

table width of top of diamond relative to widest point (43--95)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
#read csv file from directory
df = pd.read_csv('./data/diamonds.csv',sep = ',',index_col='Unnamed: 0')
```

```python
#Explore 5 rows of this data
df.head()
```

```python
#describe data with basic statistic
df.describe()
```

```python
#There are missing values in 'x','y','z' columns in value is zero
#Clean data by drop zero value
df_clean = df[df['z']>0|(df['y']>0)|(df['x']>0)]
```

```python
# transform object to numerical by LabelEncoder 
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
cut_label = enc.fit_transform(df_clean.cut)
color_label = enc.fit_transform(df_clean.color)
claity_label = enc.fit_transform(df_clean.color)
```

```python
#Transform in to dataframe
Label_enc = pd.DataFrame({'Unnamed: 0':df_clean.index,'cut_label':cut_label,'color_label':color_label,'claity_label':claity_label})
Label_enc = Label_enc.set_index('Unnamed: 0')
```

```python
Label_enc.head()
```

```python
# Merge dataframe 
df_enc = pd.concat([df_clean,Label_enc],axis=1)
```

```python
# Drop object columns
df_enc = df_enc.drop(['cut','color','clarity'],axis=1)
```

```python
df_enc.describe()
```

```python
df_enc.corr()['price']
```

```python
X = df_enc.drop(['price'],axis=1)
y = df_enc['price']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train_nor = scale.fit_transform(X_train)
X_test_nor = scale.fit_transform(X_test)
```

```python
X_train_nor = pd.DataFrame(X_train_nor,columns = X_train.columns)
X_test_nor = pd.DataFrame(X_test_nor,columns = X_train.columns)
```

### Algorithm Evaluation

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor

models = []
results = []
names = []
models.append(('linear regression',LinearRegression()))
models.append(('lasso',Lasso()))
models.append(('ridge',Ridge()))
models.append(('Tree regression',DecisionTreeRegressor()))

for name,model in models:
    result = cross_val_score(model,X_train_nor,y_train)
    results.append(result)
    names.append(name)
    print(f'the result of {name} is {result.mean()} with {result.std()}')
```

```python
lasso_model = Lasso()
lasso_model.fit(X_train_nor,y_train)
y_lasso_model = lasso_model.predict(X_test_nor)
mae_lasso = mean_absolute_error(y_lasso_model,y_test)
mse_lasso = mean_squared_error(y_lasso_model,y_test)
print(f'MAE : {mae_lasso}')
print(f'MSE : {mse_lasso}')
```
