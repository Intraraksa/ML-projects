```python
df = data[['Category','Date','X','Y']]
gta = df[df['Category']=='VEHICLE THEFT']
gta['Day'] = gta['Date'].apply(lambda day : day.strftime("%a"))
```

```python
gta_weekend = pd.Series([gta[gta['Day']=='Sun']] and [gta[gta['Day']=='Sat']])
```

```python
san = folium.Map((37.773972,-122.431297),zoom_start=12)
HeatMap(list(zip(gta['Y'],gta['X'])),radius=14).add_to(san)
san
```
