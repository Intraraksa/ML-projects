```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('voted-kaggle-dataset.csv')
```

```python
data.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
import re
```

```python
data = data.dropna(axis = 0,subset = ['Description'])
```

```python
def preprocess(text):
    word = re.sub('[^A-za-z]',' ',text)
    word = word.lower()
    word = word.split()
    word = [t for t in word if t not in sw]
    word = [WordNetLemmatizer().lemmatize(t) for t in word]
#     word = ' '.join(word)
    return word
```

```python
data['Description'][11]
```

```python
sw = stopwords.words('english')
text = data['Description'].apply(preprocess)
```

```python
data.info()
```

```python
import gensim
```

```python
dictory = gensim.corpora.Dictionary(text)
```

```python
doc2bow = [dictory.doc2bow(t) for t in text]
```

```python
model = gensim.models.LdaModel(corpus = doc2bow,id2word=dictory,num_topics=15,passes=6)
```

```python
model.top_topics(doc2bow)
```
